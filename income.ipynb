```python
# Import data file from CSV to dataframe

path_income_data = "Data/income_data.csv"

raw_income_data = pd.read_csv(path_income_data)
raw_income_data.head()

```

```python
#get column names
raw_income_data.columns
```

```python
#check data types
raw_income_data.dtypes
```

```python
#drop unnecessary columns
raw_income_data=raw_income_data.drop(['Flag Codes','Flags','TIME','REGIONTYPE','Geography Level','SEX_ABS','Sex','AGE','Age','Census year','STATE'],1)
raw_income_data.head()
```

```python
#Keep only WA data
income_data_wa=raw_income_data[raw_income_data["State"] =="Western Australia"]
income_data_wa=income_data_wa.drop(["State"],1)
income_data_wa.head()
```

```python
#Rename income column
income_data_wa=income_data_wa.rename(columns={'Total Personal Income - weekly': "Income"})
income_data_wa
```

```python
#check income values
list(income_data_wa.Income.unique())
```

```python
#make nil and negative incomes 0

income_data_wa["Income"]=income_data_wa["Income"].replace('Negative income ', '0-0 (0-0)')
income_data_wa["Income"]=income_data_wa["Income"].replace('Nil income ', '0-0 (0-0)')
income_data_wa["Income"]=income_data_wa["Income"].replace('$3,000 or more ($156,000 or more) ', '3000-3000 (156000-156000)')
list(income_data_wa.Income.unique())
```

```python
#drop not stated and TOTAL
income_data_wa=income_data_wa[income_data_wa["Income"] !="Not stated "]
income_data_wa=income_data_wa[income_data_wa["Income"] !="Total"]
list(income_data_wa.Income.unique())
```

```python
#get annual income range
new = income_data_wa["Income"].str.split("(", n = 1, expand = True) 

# making separate column from new data frame 
income_data_wa["Annual_Income"]= new[1]



#drop , and ()
income_data_wa["Annual_Income"]=income_data_wa["Annual_Income"].str.replace(")","") 
income_data_wa["Annual_Income"]=income_data_wa["Annual_Income"].str.replace(" ","") 
income_data_wa["Annual_Income"]=income_data_wa["Annual_Income"].str.replace(",","") 
income_data_wa["Annual_Income"]=income_data_wa["Annual_Income"].str.replace("$","") 
list(income_data_wa.Annual_Income.unique())
```

```python
#split annual income into lower and upper, make float or int
new=income_data_wa['Annual_Income'].str.split("-", n = 2, expand = True)
income_data_wa['Annual_Income_Lower']= new[0].astype(float)
income_data_wa['Annual_Income_Upper']= new[1].astype(float)
income_data_wa.head()
```

```python
#find mid point annual income, multiply by people in that range (value)
income_data_wa['Annual_Income_Midpoint']=(income_data_wa['Annual_Income_Lower']+income_data_wa['Annual_Income_Upper'])/2
income_data_wa['Annual_Income_Midpoint_Total']=(income_data_wa['Annual_Income_Midpoint']*income_data_wa['Value'])
```

```python
#fix region column 
new = income_data_wa["Region"].str.split("(", n = 1, expand = True) 

#  column from new data frame 
income_data_wa["Region"]= new[0]




#drop extra rows
sum_income_data_wa=income_data_wa.drop(['Annual_Income_Lower','Annual_Income_Upper','Annual_Income_Midpoint','Income','LGA_2016','Annual_Income','INCP_2016'],1)



sum_income_data_wa.head()
```

```python

mean_income_data_wa=sum_income_data_wa.groupby("Region").sum().round(2)
mean_income_data_wa
```

```python
#import region/district

path_region_district= "Data/region_by_district_wa.csv"

region_district = pd.read_csv(path_region_district)
region_district.head()

```

```python
#merge to put income with district
income_with_district=pd.merge(mean_income_data_wa,region_district,left_on='Region',right_on='Region')
income_with_district.head()
```

```python
#find total average income per district
income_by_district=income_with_district.groupby("District").sum()
income_by_district['Mean_Income']=(income_by_district['Annual_Income_Midpoint_Total']/income_by_district['Value']).round(2)

income_by_district.head()
```

```python
income_by_district=income_by_district.drop((["Value"]),1)
income_by_district=income_by_district.drop((['Annual_Income_Midpoint_Total']),1)
income_by_district
```
